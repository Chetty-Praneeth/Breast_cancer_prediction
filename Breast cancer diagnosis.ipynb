**Training model**

```python
prediction_vars = ['radius_mean','perimeter_mean','area_mean','compactness_mean','concavity_mean','concave points_mean','radius_se','perimeter_se','area_se','radius_worst','perimeter_worst','compactness_worst']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[prediction_vars],df['diagnosis'],test_size=0.15,random_state=1)
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'max_depth' : (1,2,3,4),'n_estimators' : (10,50,100,500)}
best_model =  GridSearchCV(model,parameters)
```

```python
best_model.fit(X_train,y_train)
```

```python
best_model.best_params_
```

```python
model.fit(X_train,y_train)
```

```python
score = model.score(X_test,y_test)
```

**Grid Search**

```python
y_predicted  = best_model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_predicted,y_test)
cm
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precision = precision_score(y_test,y_predicted)
print(f"The precision is: {precision:.2f}")
recall = recall_score(y_test,y_predicted)
print(f"The recall is: {recall:.2f}")
```
